```python
#columns_to_check = ['customer_children', 'customer_relationship', 'customer_self_employed', 'customer_occupation_code', 'customer_education',  'customer_since_all', 'customer_since_bank', 'customer_gender', 'customer_birth_date', 'customer_postal_code']
columns_to_check = [c[3:] for c in train1.columns.to_list()]

ok = []

for c in columns_to_check:
    m1 = 'm1_' + c
    m2 = 'm2_' + c
    m3 = 'm3_' + c
    
    if train[m1].equals(train[m2])  and train[m1].equals(train[m3]):
        ok.append(c)
        
print(ok)
```

```python
for c in ok:
    m1 = 'm1_' + c
    m2 = 'm2_' + c
    train.drop(m1, axis=1, inplace=True)
    train.drop(m2, axis=1, inplace=True)
    
train.head()
```

```python
for c in ok:
    m3 = 'm3_' + c
    train.rename(columns={m3: c}, inplace=True)
    
train.head()
```

```python
# Check missingness
m = {}

for c in train.columns:
    v = train[c].isnull().sum()
    if v != 0:
        m[c] = v
    
print(m)
```

```python
train[['birth_year', 'birth_month']] = train['customer_birth_date'].str.split('-', 1, expand=True)
train.columns
```

```python
train.drop('customer_birth_date', axis=1, inplace=True)
train.head()
```

```python
train[train['birth_year'].astype(int) < train['customer_since_bank_year'].astype(int)]
```

```python
train[['customer_since_all_year', 'customer_since_all_month']] = train['customer_since_all'].str.split('-', 1, expand=True)
train[['customer_since_bank_year', 'customer_since_bank_month']] = train['customer_since_bank'].str.split('-', 1, expand=True)
```

```python
train.drop('customer_since_all', axis=1, inplace=True)
train.drop('customer_since_bank', axis=1, inplace=True)
train.head()
```
